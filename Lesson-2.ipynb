```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#1
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
plt.plot(x,y, color = 'r', linewidth = 4.0)
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

```python
plt.scatter(x,y, color = 'g', linewidth = 2.0)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#2 
t = np.linspace(0, 10, 51)
f = np.cos(t)
plt.plot(t, f, 'g', linewidth = 2.0)
plt.title('График f(t)', fontsize=14)
plt.xlabel('t', fontsize=14), plt.ylabel('f', fontsize=14)
plt.xlim(0.5, 9.5), plt.ylim(-2.5, 2.5)
plt.grid()
plt.show() 
```

```python
#3
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)

fig = plt.figure(figsize= (8, 6))

ax1 = plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.grid()
plt.title ('График y1', fontsize=14, color='g')

ax2 = plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.grid()
plt.title ('График y2', fontsize=14, color='g')

ax3 = plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.grid()
plt.title ('График y3', fontsize=14, color='g')

ax4 = plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.grid()
plt.title ('График y4', fontsize=14, color='g')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
#3
fig, ax = plt.subplots(nrows=2, ncols=2, sharey = True)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)


ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.grid() 

ax2.plot(x, y2)
ax2.set_title("График y2")
ax2.grid() 

ax3.plot(x, y3)
ax3.set_title("График y3")
ax3.grid() 

ax4.plot(x, y4)
ax4.set_title("График y4")
ax4.grid() 

plt.show()
```
